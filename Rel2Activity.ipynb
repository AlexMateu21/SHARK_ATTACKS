```python
import pandas as pd
url = ("https://www.sharkattackfile.net/spreadsheets/GSAF5.xls")
sharkattacks_data = pd.read_excel(url)
print(sharkattacks_data.head(10))
```

```python
# 1) Keep provoked only
provoked = sharkattacks_data[
    sharkattacks_data["Type"].str.contains("Provoked", case=False, na=False)
].copy()
```

```python
# 2) What activity is most linked to provoked attacks?
provoked_activity = provoked["Activity"].value_counts().head(15)
display(provoked_activity)
```

```python
# 3) When do provoked attacks happen most? (using Time as recorded text)
provoked_time = provoked["Time"].value_counts().head(15)
display(provoked_time)

display(provoked)
```
